```python
import numpy as np 
import pandas as pd
df=pd.read_csv('train.csv')
df.head(10)
```

```python
df.drop(['id'],axis=1,inplace=True)
```

```python
df.head()
```

```python
x=df['tweet']
y=df['label']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words="english",max_df=.90,min_df=0.10)
xtr=cv.fit_transform(x)

from sklearn.naive_bayes import MultinomialNB
naive=MultinomialNB()
naive.fit(xtr,y)
```

```python
dft=pd.read_csv('test.csv')
dft.drop(["id"],axis=1,inplace=True)
dft.head(10)
```

```python
xts=dft['tweet']
xts=cv.transform(xts)
```

```python
opt=naive.predict(xts)
```

```python
for i in opt:
    print(i)
```
